```python
df2=df.str.extractall(r'((\d?\d)[/-](\d?\d)[/-](\d\d\d\d))')# extracted day/month/year(4 digit)
df2
df21=df2.rename({0:'Totaldate',1:'Month2',2:'Day2',3:'Year2'},axis=1)
df21[['Day2','Month2','Year2']]
```

```python
df3=df.str.extractall(r'((\d?\d)[/-](\d?\d)[/-]((?:\d{2})(?!\d+)))')# extracted day/month/year(2 digit)
df31=df3.rename({0:'Totaldate',1:'Month3',2:'Day3',3:'Year'},axis=1)
df31
```

```python
grades=[]
for row in df31.Year:
    if int(row)>10:
        grades.append("19"+str(row))      
    else:
        grades.append("20"+str(row))
            
df31['Year3']=grades
#df31['Year1']
df31[['Day3','Month3','Year3']]
```

```python
df4=df.str.extractall(r'((?<![\d+/])(\d?\d)[/-]((?:\d{4})(?!\d+)))')# extracted with month/year
df4
df41=df4.rename({0:'Totaldate',1:'Month4',2:'Year4'},axis=1)
df41
```

```python
grades1=[]
for row in df41.Month4:
    if int(row)<350:
        grades1.append('01')      
    else:
        grades1.append("20")
            
df41['Day4']=grades1
df41[['Day4','Month4','Year4']]
```

```python
df5=df.str.extractall(r'(?:((\d?\d)\s?([^\n][Jan|Feb|Mar|Apr|May|Jun|Aug|Sep|Oct|Nov|Dec]\w+)\s?\,?(\d{4})))')# extracted with month/year
df5
df51=df5.rename({0:'Totaldate',1:'Day5',2:'Month5',3:'Year5'},axis=1)
df51[['Day5','Month5','Year5']]
```

```python
month_conv=[]
for row in df51.Month5:
    if str(row) =='Jan'or str(row) =='January':
        month_conv.append('01')
    elif str(row)==str('Feb') or str(row) =='February':
        month_conv.append('02')
    elif str(row)==str('Mar') or str(row) =='March':
        month_conv.append('03')
    elif str(row)==str('Apr') or str(row) =='April':
        month_conv.append('04')
    elif str(row)==str('May') or str(row) =='May':
        month_conv.append('05')
    elif str(row)==str('Jun') or str(row) =='June':
        month_conv.append('06')
    elif str(row)==str('Jul') or str(row) =='July':
        month_conv.append('07')
    elif str(row)==str('Aug') or str(row) =='August':
        month_conv.append('08')
    elif str(row)==str('Sep') or str(row) =='September':
        month_conv.append('09')
    elif str(row)==str('Oct') or str(row) =='October':
        month_conv.append('10')
    elif str(row)==str('Nov') or str(row) =='November':
        month_conv.append('11')
    else:
        month_conv.append('12')
    
df51['Month51']=month_conv
df51[['Day5','Month5','Month51','Year5']]
```
