```python
import pandas as pd
import numpy as np
df=pd.read_csv('titanic.csv')
df.isna().sum()
```

```python
inputs=df.drop(['Survived','PassengerId','Name','Age','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
target=df.Survived
inputs.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex=LabelEncoder()
inputs['Sex_n']=le_Sex.fit_transform(inputs['Sex'])
```

```python
inputs.drop(['Sex'],axis=1,inplace=True)
inputs
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
model=tree.DecisionTreeClassifier(criterion='entropy')

X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.25,random_state=1)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
