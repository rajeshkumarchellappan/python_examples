```python
import pandas as pd

doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

df = pd.Series(doc)
df
```

```python
df['dateextract1']=df.str.findall(r'\d+[/-]\d+[/-]\d{2,4}')
#f['dateextract']=df['text'].str.findall(r'(\d{4})')
#df['dateextract']=df['text'].str.findall(r'(\d\d\d\d)')
#df.text.str.count(r'(\d+[/-]\d+[/-]\d+)').sum()
df['dateextract1']
```

```python
df['dateextract2']=df.str.replace('(\d{5})','',regex=True)
df['dateextract3']=df['dateextract2'].str.findall(r'\d+[/-]\d+[/-]\d{2,4}')
df['dateextract3']
```
